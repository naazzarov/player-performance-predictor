```python
import pandas as pd

# Load the dataset
df = pd.read_csv("players.csv")

# Display the first few rows
df.head()
```

```python
# Show info about dataset
df.info()

# Check for missing values
df.isnull().sum()
```

```python
# Strip whitespaces from column names
df.columns = df.columns.str.strip()

# Strip whitespaces from object (string) columns
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

# Convert 'Seasons Ratings' to float (in case it was read as object)
df['Seasons Ratings'] = pd.to_numeric(df['Seasons Ratings'], errors='coerce')

# Recheck for any NaNs
df.isnull().sum()
```

```python
# Show cleaned data
df.head()

# Show summary stats
df.describe(include='all')
```
